```python
# ログ用csvファイルの読み込み(壁無し)

path = "C:/Users/qtfdl/Desktop/log.csv"
csv_data = np.loadtxt(path, delimiter=",", dtype = float, skiprows=1)

TIMESTAMP_COL = 0
WALL_L_COL = 28
WALL_R_COL = 29
THREE_SECTION_AHEAD_DIST_COL = 52

timestamp = csv_data[:, TIMESTAMP_COL]
wall_r = csv_data[:, WALL_L_COL]
wall_l = csv_data[:, WALL_R_COL]
three_section_ahead_dist = csv_data[:, THREE_SECTION_AHEAD_DIST_COL]

wall_corner_threshold_on_l = 250
wall_corner_threshold_off_l = 190
wall_corner_threshold_on_r =  250
wall_corner_threshold_off_r = 190
  
```
